```python
import pandas as pd
```

```python
survey1 = pd.read_csv(r"C:\Users\akash\Desktop\survey_results_public.csv")
survey1
```

```python
survey1.columns
```

```python
survey=survey1.loc[:,['Country','LanguageHaveWorkedWith','ConvertedCompYearly']]
survey
```

```python
survey.info()
```

```python
survey["Country"].fillna(survey["Country"].mode()[0],inplace =True)
survey.info()
```

```python
survey['LanguageHaveWorkedWith']=survey['LanguageHaveWorkedWith'].str.split(';')
```

```python
survey=survey.explode("LanguageHaveWorkedWith")
```

```python
survey
```

```python
survey['LanguageHaveWorkedWith'].fillna(survey['LanguageHaveWorkedWith'].mode()[0],inplace=True)
```

```python
survey.info()
```

```python
survey['ConvertedCompYearly'].fillna(survey['ConvertedCompYearly'].median(),inplace =True)
```

```python
survey.info()
```

```python
survey2=survey['LanguageHaveWorkedWith']=='Python'
av=survey[survey2]
av
```

```python
survey5 = av.groupby('Country')
```

```python
survey6 = pd.DataFrame(survey5['ConvertedCompYearly'].mean())
survey6
```

These are average salary of python developers in each country

There were a lot of null values in the dataset , we fill the null value with the help of measures of central tendency.we use mode for filling country and LanguageHaveWorkedWith and median for ConvertedCompYearly
